Our data includes annual number of deaths per country or region by cause of death. <br>
We cleaned our data by considering which means would correlate with country and which would not. We grouped by mean and country for every category except Fire and Drowning. We also removed the "Code" and "Execution" categories because they lacked sufficient data to draw conclusions based on a world wide data representation. Then, we renamed all columns to reflect the category without (Deaths) as we could create labels for our charts that reflected that the values represented were the number of deaths. This helped to remove clutter from our visualizations.

Below, you will find our 7 visualizations: bar chart, boxplot, line plot, scatter plot, histogram, heat map, and pie chart. Each chart will have an explanation for how the data was manipulated and explaining the chart.

```
'''
*Bar Chart Strategy:

We wanted to study the events of terrorism across the years with a visual aid. The main goal of this visualization was 
to notice if regions with reduced political stability or a lack of governance thereoff had an increased incidence of 
terrorism and if that correlation could be determined to be causation.

*Bar Chart Description:

The bar graph shown above uses a stacked visual approach to better visualize the large amount of data from across the years
1990 to 2017. The death counts by terrorism increased dramatically from 2004 and reached the peak in 2014, which showed that
many regions in the world are more and more unsecured. 

We have focussed mainly on the regions which the pivot helped show had the largets incidents of deaths from terrorism. We do
notice a trend of the highest deaths being form areas with reduced governance or a lack of political stabiltiy. 
An interesting observation we found was that the areas geographically linked with these areas showed an increase or high 
incidence of deaths due to terrorism.
'''

dv_bar=death_data.groupby(['Year','Entity'])['Terrorism'].sum().unstack()

dv1_bar=dv_bar[['Middle East & North Africa','South Asia','Iraq','Sub-Saharan Africa','Afghanistan','Pakistan']]
axes=dv1_bar.plot.bar(stacked=True, fontsize=15, figsize = (12,7))
axes.legend(loc=2,bbox_to_anchor=(1, 1))
axes.set_xlabel("Year", size = 15)
axes.set_ylabel("Number of Deaths", size = 15)
axes.set_title('Deaths by Terrorism in top 6 Regions/Countries', size = 20)
```

```
'''
*Box Plot Strategy:

With the box plot we wanted to study the incidences of suicides and homicides with their distribution among their own data 
sets over the years.
We wanted to determine if the normalized trend of homicides and suicides has an average median or is consistenlty low over
the years. This could explain certain spikes in deaths during specific years.

*Box Plot Description:

The Box plot was drawn using the data for deaths from homicides and suicides across the years. We noticed the following 
points from the visualization:

1)The median for both the plots are on the lower end of the plot
2)Both the plots are postiviely skewed or skewed right
3)Very few outliers and always on the lower end
4)Death counts by suicide are more than triple to death counts by homicide, which suggests that the mental health of 
human being should be seriously taken care of. 

We felt that this showed that cases of suicides and homicides are usually on the lower end with spikes in certain years only.


'''
dv_box=death_data.groupby('Year')['Suicide','Homicide'].sum()

dv_box

axes=dv_box.plot.box(grid=True, rot=45, fontsize=15, figsize = (12,7))
axes.set_xlabel("Type of Deaths", size = 15)
axes.set_ylabel("Number of Deaths", size = 15)
axes.set_title('Distribution of Deaths by Suicide and Homicide', size = 20)

```

```
#Creating pivot table to sum deaths by malaria by year and entity
dv=death_data.groupby(['Year','Entity'])['Malaria'].sum().unstack()
dv
```

```
'''
*Line Graph Strategy:

For the line graph we wanted to study the following:
1)Instances of malaria in the most affected regions in the world
2)The changes of its occurrence over the years
3)The effect of global aid in these countries and how it has helped

*Line Graph Description:

The line graph we were able to visualize the 5 largest regions of the world with malaria. We found the following:

1)Malaria grew slowly but consistenly over the years from 1990
2)It peaked in 2004 after which there has been a steady decline in cases in the most prone areas
3)The world trend also conforms to the trend in the most afflicted areas
4)The number of deaths by malaria started a steep decline after the funds pledged by organizations specifiaclly the Bill 
and Melinda Gates foundation.

'''

dv1=dv[['World','Sub-Saharan Africa','Western Sub-Saharan Africa','Eastern Sub-Saharan Africa','Central Sub-Saharan Africa','South Asia']]
dv1
plt.plot(dv1)
plt.legend(dv1.columns,loc='lower center',framealpha=1,bbox_to_anchor=(1.30, 0))
plt.xlabel('Years')
plt.ylabel('Total number of deaths')
plt.title('Distribution of largest Malaria deaths over the years', size = 15)

```

```
'''
*Scatter Plot Strategy

With the Scatter Plot we wanted to study the follwing:
1)The growth of the AIDS pandemic in the years 1990 to 2017
2)The peak incidences of AIDS in the world as a whole
3)The effect of better health care on the loss of life due to AIDS 

*Scatter Plot Description

With the Scatter Plot of AIDS deaths across the world all plotted agaisnt the years we could infer the following:
1)The AIDS pandemic grew steadily form 1990 until 2005 when it peaked in most countries
2)With improved health care and anti retrovial drugs the deaths have seen a sharp decline after 2005
3)This aso correlates to a PSA campaign which recommended the use of condoms and safe sex education among young adults

'''

plt.scatter(death_data['Year'],death_data['HIV/AIDS'])
plt.xlabel('Years')
plt.ylabel('Total number of deaths (Millions)')
plt.title('Distribution of HIV/AIDS deaths over the years', size= 15)
```

```
'''
*Pie Chart Strategy:

The goal of the pie chart was to analyse the following data:
1)Proportions of the largest causes of deaths in the world
2)Identifying the largest treatable causes and studying better approaches to solving them
3)See any correlation among the largest causes to help suggest solutions among them

*Pie Chart Description:

The Pie Chart we were able to visualize utilized the number of deaths from the 10 largest causes in the dataset cumulative 
from ten years of 1990 to 2017. It accomplished the following:
1)It helped showcase that cardiovascular disease is the largest cause of death in the world
2)It also showcased cancer to be a major cause of death in the world
3)It helped establish that most major reasons for death are not correlated in an openly obvious way

'''

#Create pivot table to plot the heat map: deaths by year
df_pie=death_data.sum()[['Cardiovascular diseases','Cancers','Respiratory diseases','Lower respiratory infections','Neonatal disorders','Digestive diseases','Diarrheal diseases','Dementia','Tuberculosis']]

df_pie.plot.pie(autopct='%1.1f%%',shadow=True,title='Largest causes of death', figsize = (12,7))

```

```
'''
*Histogram Strategy:

With the histogram we were hoping to visualize the incidences of mmeningitis related deaths in the 10 most affected nations.
Since Meningitis usually has underlying causes and issues which trigger it, the hope was to learn the frequency of deaths in
these nations and identify if there was a corellation among the general healthcare level and economy of the country against 
the incidence of the disease

*Histogram Description:

The histogram showed that meningitis death are right skewed with death counts are usually at the level from 455,000 to 460,000.
The high frequency of deaths in meningitis occurs in countries with strained health care and economic resources. It also 
corellated with regions which had high incidences of tropical diseases and possible underlying conditions.

'''

hist_df=death_data.groupby(['Year','Entity'])['Meningitis'].sum().nlargest(10)
hist_df

plt.hist(hist_df,density = 1,color ='green', alpha = 0.7) 
plt.title('Distribution of death related to Meningitis', size = 15)   
plt.xlabel('Deaths', size = 10)
plt.ylabel('Frequency of Occurence', size = 10)
```

```
#Creating pivot table to plot the heat map: deaths by year
pivot = death_data.pivot_table(index=['Year'], values=['Meningitis','Lower respiratory infections','Intestinal infectious diseases','Protein-energy malnutrition','Terrorism','Cardiovascular diseases','Dementia','Kidney disease','Respiratory diseases','Liver diseases','Digestive diseases','Hepatitis','Cancers','Parkinson disease','Fire','Malaria','Drowning','Homicide','HIV/AIDS','Drug use disorders','Tuberculosis','Road injuries','Maternal disorders','Neonatal disorders','Alcohol use disorders','Natural disasters','Diarrheal diseases','Temperature exposure','Nutritional deficiencies','Suicide','Conflict','Diabetes','Poisonings'], aggfunc='sum')                            

print (pivot)
```

```
'''
*Heat Map Strategy:

In order to see the change of total death counts by various reasons from 1990 to 2017, we decided to use the heat map. 
Heat map will allow us to decide which causes of death were becoming more popular by the time and which causes of death were
better controlled over the years. It also shows us the popularity of each death cause compared to the others. Hence, the 
heat map will help to verify if any improvements to control the death causes were efficient and suggest what causes needed 
to be focused in the future.

From the data, we created a pivot table in Python to calculate deaths by different causes and by different years. The rows 
of pivot table are different years from 1990 to 2017, the columns are different causes of death and the values are death counts.

From the pivot table, we plotted the heat map. The x-axis of the map is different causes of deaths and y-axis is the years 
from 1990 to 2017. The green color shows magnitude of death count, with 8 being the highest level and is shown with darkest 
color, 0 being the lowest level and is shown with lightest color. If the color of a certain kind of death cause turns darker
by the time, it means that death cause is becoming more popular and vice versa. The darkness of color also helps us to 
compare the popularity of different death causes at a certain time. 


*Heat Map Description:

Two main reasons for deaths over the years are cardiovascular diseases and cancers. Cardiovascular diseases deaths increased
drastically from 1990 to 2007 due to increased population and the rise of unhealthy lifestyle. Cancers deaths also increased
over the years but the degree of increase is not apparent. 

The next three main causes of deaths are respiratory diseases, lower respiratory infections and neonatal disorders. However,
there was a slight decrease in deaths of those causes, mainly due to the development of healthcare systems.

Diarrheal diseases and digestive diseases are the next causes of deaths. While digestive deaths were increasing gradually 
from 1990 to 2017, deaths by diarrheal diseases decreased since around 2000. The increased deaths by digestive diseases 
might be because of the unhealthy and unbalanced eating habit of population, while the decrease in deaths by diarrheal could
be explained by the improvement in hygiene practices and healthcare system.

HIV/AIDS is also among the list of popular causes of deaths. The heat map showed that deaths by HIV/AIDS increased greatly
from 1997 to around 2000 mainly because of unprotected sex and share of drug needles. After 2010, the surge of HIV/AIDS was 
gradually controlled thanks to worldwide campaigns to prevent the diseases and introduction of new medicines to relieve its 
symptoms.   

Dementia is another popular cause of deaths. The heat map showed that deaths by dementia increased from 2002 onward. This 
could be explained by the aging population as dementia is closely related to people’s age.

Other causes of deaths are minor and remained stable over the years.

''' 
     
fig, ax = plt.subplots(figsize=(12,7))
im = ax.imshow(pivot, cmap="Greens")
fig.colorbar(im, ax=ax)

ax.set_xticks(range(len(pivot.columns)))
ax.set_yticks(range(len(pivot.index)))
ax.set_xticklabels(pivot.columns, rotation=90)
ax.set_yticklabels(pivot.index)
ax.set_xlabel("Disease")
ax.set_ylabel("Year")
ax.set_title("Deaths by different reasons from 1990 to 2017", size=20)

plt.show()


```

Conclusion

The visualization of data set drew a clear picture on main causes and distribution of death in different countries and regions over the years. It also helps us to see the trend of deaths throughout the years. The bar chart has shown that terrorism is increasing by the years and is higher at regions/countries with unstable political system or reduced governance.

From the box plot, we noticed that the death counts by suicide is higher than that of homicide, showing the importance of maintaining good mental health for people. Both plots are right skewed with very few outliers, which means that both suicide and homicide are usually on the lower end with spikes in certain years only.

The line chart visualized a steady growth of deaths by malaria from 1990 to 2004 then a gradual decrease afterward in most areas. This decline can be explained by efforts and funding activities to help repel the diseases.

Through the scatter plot, we can see that the AIDS pandemic grew steadily from 1990 to 2005. The propaganda along with improved healthcare system and introduction of medicine had a positive impact on the AIDS death growth after 2005.

The pie chart showed us that cardiovascular and cancers are among top reasons for deaths, hence, a healthy lifestyle and diet are very important to people’s longevity.

The histogram visualized a large portion of deaths by meningitis are between 455,000 and 460,000 for each country. This high frequency is due to limit in healthcare system and economic conditions in many countries.

From the heat map, we can see that cardiovascular, cancer, digestive and dementia related deaths are increasing by the years, mainly due to unhealthy lifestyle and aging population. In contrast, respiratory diseases and HIV/AIDS related deaths are decreasing thanks to development of healthcare system and worldwide campaigns.

The insights gained from our charts are key to evaluate development of healthcare system and efforts to repel the diseases and avoid certain deaths. They also help us to identify which death causes are dangerous and need more control in the future.
